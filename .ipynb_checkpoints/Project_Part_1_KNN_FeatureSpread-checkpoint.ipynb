```python
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
```

```python
breast_cancer_data
```

```python
X,y = dataset[0],dataset[1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
len(X_train) == len(y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

# KNN Classifier...choosing best N

```python
best_score = 0
best_k = 0
for i in range(1,25):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    knn_score = knn.score(X_test,y_test)
    if knn_score > best_score:
        best_score = knn_score
        best_k = i
```

```python
best_score
```

```python
best_k
```

## Cross-validate

### Small dataset can handle multi-fold-cross validation

```python
best_cv_score = 0
best_cv_k = 0
model = None 
for k in range(1,25):
    knn_cv = KNeighborsClassifier(n_neighbors=k)
    scores_cv = cross_val_score(knn_cv, X_train, y_train, scoring='accuracy',cv=10)
    mean_scores_cv = scores_cv.mean()*100
    
    print('Mean accuracy with 10-fold cross-validation is %s' % round(mean_scores_cv,2),' k=',k)

    if mean_scores_cv > best_cv_score:
        best_cv_score = mean_scores_cv
        best_cv_k = k
        model = knn_cv
```

```python
best_cv_score
```

```python
best_cv_k
```
